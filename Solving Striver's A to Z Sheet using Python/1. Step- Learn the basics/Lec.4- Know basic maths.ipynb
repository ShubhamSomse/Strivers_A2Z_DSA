```python


def reversenum(num):
    ans = 0
    while num > 0:
        digit = num % 10          # extract last digit
        ans = ans * 10 + digit    # build reversed number
        num = num // 10           # remove last digit
    return ans
num = int(input())
print(reversenum(num))
```

```python
def reversenum(num):
    ans = 0
    while num > 0:
        rem = num % 10
        ans = ans * 10 + rem
        num = num // 10
    return ans

def palindrom(num):
    if num == reversenum(num):
        print("Palindrom")
    else:
        print("Not Palindrom")

num = int(input())
palindrom(num)
```

```python
def gcd(n1, n2):
    while n1 > 0 and n2 > 0:
        if n1 > n2:
            n1 = n1 % n2
        else:
            n2 = n2 % n1

    if n1 == 0:
        return n2
    else:
        return n1

n1 = int(input())
n2 = int(input())
print(gcd(n1, n2))
```

```python
def armstrong(num):
    original = num
    digits = len(str(num))
    total = 0

    while num > 0:
        digit = num % 10
        total += digit ** digits
        num //= 10

    return total == original


num = int(input())
print(armstrong(num))
```

```python
def checkdivisor(num):
    ans = []
    for i in range(1, num + 1):
        if num % i == 0:
            ans.append(i)
    return ans

num = int(input())
print(checkdivisor(num))
```

```python
#optimized
def checkdivisor(num):
    ans = []
    i = 1
    while i * i <= num:
        if num % i == 0:
            ans.append(i)
            if i != num // i:
                ans.append(num // i)
        i += 1
    return sorted(ans)

num = int(input())
print(checkdivisor(num))
```

```python
def isprime(num):
    if num <2:
        return False
    i = 2
    while  i*i <= num:
        if num % i == 0:
            return False
        i += 1
    return True

num = int(input())
print(isprime(num))
```
