```python
n = int(input("enter the rows:"))
```

```python
for i in range(n):
    for j in range(n):
        print("*", end ="")
    print()
```

```python
for i in range(n):
    for j in range(n-i+1):
        print(" ", end ="")
    for j in range(2*i-1):
        print("*", end = "")
    print("")
```

```python
for i in range(n):
    for j in range(i+1):
        print(" ", end ="")
    for j in range(2*(n-i)-1):
        print("*", end= "")
    print()
```

```python
for i in range(n):
    for j in range(i):
        print(" ", end="")
    for j in range(2*(n-1-i)+1):
        print("*", end = "")
    print()
```

```python
n = 10

for i in range(n):
    for j in range(n+i-1):
        print(" ", end ="")
    for j in range(2*(n-i)-1):
        print("*", end ="")
    print()
```
