```python
n = int(input("enter row:"))
t = int(input("enter column:"))
```

```python
for i in range(n):
    for j in range(n):
        print("*", end= "")
    print()
```

```python
for i in range(n):
    for j in range(i+1):
        print("*", end= "")
    print()
```

```python
for i in range(n):
    for j in range(i+1):
        print(j+1, end = "")
    print()
```

```python
for i in range(n):
    for j in range(i+1):
        print(i+1, end ="")
    print()
```

```python
for i in range(n):
    for j in range(n-i):
        print("*", end ="" )
    print()
```

```python
for i in range(n):
    for j in range(n-i):
        print(j+1, end= "")
    print()
```
