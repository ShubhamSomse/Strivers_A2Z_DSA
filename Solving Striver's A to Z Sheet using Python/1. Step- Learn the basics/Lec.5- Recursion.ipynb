
process of calling a same function again & again until a specific condition is met
Examples:
1.Print a Name n times
2. Print 1 to N using recursion
3. Print N to 1 using Recursion
4. Sum of first n numbers
5. Factorial of a number n
6. Reverse an array
7. Check if string is palindrom or not
8. Fibonacci Number

```python
#infinite loop
def myfun():
    while True:
        print("hello, this is loop")

myfun()
```

```python
def recursion(n):
    if n == 0:              # step 1. define base i.e. stopping condition
        return
    print("shubham")        
    recursion(n-1)
    return                  # Define recursive case

n = int(input())
recursion(n)
```

```python
def fun(i, n):
    if i > n:
        return
    print(i)
    fun(i + 1, n)

n = int(input())
fun(1, n)
```

```python
def fun(n):
    if n == 0:
        return
    print(n)
    fun(n - 1)

n = int(input())
fun(n)
```

```python
def sumofn(n):
    if n == 0:
        return 0
    return n + sumofn(n - 1)

n = int(input())
print(sumofn(n))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

n = int(input())
print(factorial(n))
```

```python
def reverse(arr, l, r):
    if l >= r:
        return
    arr[l], arr[r] = arr[r], arr[l]
    reverse(arr, l + 1, r - 1)

arr = list(map(int, input().split()))
reverse(arr, 0, len(arr) - 1)
print(arr)
```

```python
def palindrom(s , e, string):
    if s >= e:
        return True
    if string[s] != string[e]:
        return False
    return palindrom(s + 1, e - 1, string)
    
string = input()
ans = palindrom(0, len(string)-1, string)
print(ans)
```

```python
#optimized recursive Fibonacci using DP (memoization):
def fibonacci(n, dp):
    if n <= 1:
        return n
    if dp[n] != -1:          # already computed
        return dp[n]

    dp[n] = fibonacci(n-1, dp) + fibonacci(n-2, dp)
    return dp[n]

n = int(input())
dp = [-1] * (n + 1)
print(fibonacci(n, dp))
```
